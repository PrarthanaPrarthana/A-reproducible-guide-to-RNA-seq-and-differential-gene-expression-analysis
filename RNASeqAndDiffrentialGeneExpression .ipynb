In this study RNA-seq was used to characterize gene expression dysregulation in EOAD and assess possible underlying mechanisms.
Non-demented controls (NDCs) and EOAD iPSCs cases were used. The study assessed how the neurons from EOAD patients clustered with respect to NDC subjects in multi-dimensional scaling (MDS) space following correction for sex, sequencing batch, and experimental condition. This revealed no clear clustering of NDC or EOAD samples but rather a slight overlap between them. 

Differential expression analysis between two conditions revealed a small number of differentially expressed genes (DEGs) (n=14). Initially the study did not see a differnce in any gene expressed in cases or int he control. In the later stages they analysed a subset of the neuron lineage genes that mainly regulate post-mitotic neuron maturation which are NEFL, SOX2, UCHL1, TUBB3, MAP2, DCX, NEUROD1 and APOE. The study states that there were no significant Differentially exprssed genes observed in the study, however the heatmap shows that the p value for SOX2, UCHL1 and NEUROD1 is significantly less which could provide slight evidence that these genes could be differentially expressed between EOAD patients and Non-demented controls (NDCs). 

A modest difference in expression profiles between EOAD patients and non-demented control (NDC) subjects resulted in a limited number of differentially expressed genes (DEGs) in this study. Based on this analysis, the authors provide evidence that iPSC-derived neuron model systems, likely due to the loss of EOAD-associated epigenetic signatures arising from iPSC reprogramming, may not be ideal models for studying sporadic AD.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming counts is your DataFrame with expression data
counts = pd.read_csv("./resnew.csv")

# Extract all transcript IDs from the first column
transcripts_of_interest = counts[counts.columns[0]].tolist()

# Select specific transcripts and the "diffexpressed" column
data = counts.set_index(counts.columns[0]).loc[transcripts_of_interest, :]

# Extract numeric columns for PCA
numeric_data = data.drop(columns=["diffexpressed"])

# Standardize the data
standardized_data = (numeric_data - numeric_data.mean()) / numeric_data.std()

# Calculate the covariance matrix
covariance_matrix = np.cov(standardized_data.T)

# Standardize the data
standardized_data = (numeric_data - numeric_data.mean()) / numeric_data.std()

# Check for infinite or NaN values and replace them with zeros
standardized_data.replace([np.inf, -np.inf, np.nan], 0, inplace=True)

# Calculate the covariance matrix
covariance_matrix = np.cov(standardized_data.T)

# Check for infinite or NaN values in the covariance matrix
if np.any(np.isnan(covariance_matrix)) or np.any(np.isinf(covariance_matrix)):
    raise ValueError("Covariance matrix contains NaN or inf values. Check your data for missing or extreme values.")

# Calculate the eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues and eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Project the data onto the principal components
pc_data = standardized_data.dot(eigenvectors)

# Create a PCA plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pc_data.iloc[:, 0], y=pc_data.iloc[:, 1], hue=data["diffexpressed"], palette="coolwarm")
plt.title("PCA Plot of Significant Transcripts")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Expression")
plt.show()
```
